```python
# export csv files for creating ics files and sparkle events import

from datetime import datetime, timezone

def get_abstract_times(dfp1, df_sched):
    df_abstacts = pd.DataFrame(columns = ['Submission No','Title', 'Speaker', 'Moderator', 'Date Start', 'Time Start', 'Date End','Time End', 'Location', 'TimeZoneBlock', 'Categories', 'UID', 'EVENTID'])    
    for index, row in dfp1.iterrows():
        for tz in df_sched[df_sched['Title'] == row['session']]['TimeZoneBlock'].values:            
            df_selected = df_sched.loc[(df_sched['Title'] == row['session']) & (df_sched['TimeZoneBlock'] == tz)]
          
            try: 
               firstAuthorName = row['First Name'].encode('latin1').decode('utf-8') + ' ' + row['Last Name'].encode('latin1').decode('utf-8'), 
            except:
               firstAuthorName = row['First Name'].encode('utf-8').decode('utf-8') + ' ' + row['Last Name'].encode('utf-8').decode('utf-8'), 
                
            df_abstacts = df_abstacts.append({
                  'Submission No': row['Submission No'], 
                  'Title': row['Abstract Title'], 
                  'Speaker': ''.join(firstAuthorName),
                  'Moderator': "",                               
                  'Date Start': df_selected['Date Start'].values[0],
                  'Time Start': df_selected['Time Start'].values[0],
                  'Date End': df_selected['Date End'].values[0],
                  'Time End': df_selected['Time End'].values[0],
                  'Location': 'poster' + str(row['Submission No']),
                  'TimeZoneBlock': tz,
                  'Categories': 'Poster', 
                  'UID': "", 
                  'EVENTID': 'poster' + str(row['Submission No']) + tz}, 
                ignore_index=True)
    return df_abstacts

def get_posters_for_sparkle(dfp1, df_sched):
    df_abstacts = pd.DataFrame(columns = ['location', 'name', 'description', 'start', 'duration', 'eventId'])
    for index, row in dfp1.iterrows():
        for tz in df_sched[df_sched['Title'] == row['session']]['TimeZoneBlock'].values:
            df_selected = df_sched.loc[(df_sched['Title'] == row['session']) & (df_sched['TimeZoneBlock'] == tz)]
            
            try: 
                firstAuthorName = row['First Name'].encode('latin1').decode('utf-8') + ' ' + row['Last Name'].encode('latin1').decode('utf-8'), 
            except:
                firstAuthorName = row['First Name'].encode('utf-8').decode('utf-8') + ' ' + row['Last Name'].encode('utf-8').decode('utf-8'), 
            
            df_abstacts = df_abstacts.append({                  
                'location': 'poster' + str(row['Submission No']),
                'name': row['Abstract Title'],
                'description': ''.join(firstAuthorName) + ', Abstract ' + str(row['Submission No']),
                'start': int(datetime.strptime(str(df_selected['Date Start'].values[0]) + 'T' + str(df_selected['Time Start'].values[0]).zfill(6) + 'Z', '%Y%m%dT%H%M%SZ').timestamp()),
                'duration': int((datetime.strptime(str(df_selected['Date End'].values[0]) + 'T' + str(df_selected['Time End'].values[0]).zfill(6) + 'Z', '%Y%m%dT%H%M%SZ') - datetime.strptime(str(df_selected['Date Start'].values[0]) + 'T' + str(df_selected['Time Start'].values[0]).zfill(6) + 'Z', '%Y%m%dT%H%M%SZ')).total_seconds()/60),
                'eventId': 'poster' + str(row['Submission No']) + tz},
                ignore_index=True)
    return df_abstacts

df_abstracts_all = get_abstract_times(dfp, df_ph)
df_abstracts_all.to_csv('OHBM2021_schedule_posters_all.csv',index=False)

df_abstracts_for_sparkle_all = get_posters_for_sparkle(dfp, df_ph)
df_abstracts_for_sparkle_all.to_csv('OHBM2021_schedule_posters_all_for_sparkle.csv',index=False)
```
